```python
#MONITOREO DE LLENADO
from google.colab import drive
import os
import shutil
drive.mount('/content/drive',force_remount=True)
```

```python
def img_scp_mt(ruta_general):
    #entramos a las generales
    with os.scandir(ruta_general) as ficheros:
        carpetas = [fichero.name for fichero in ficheros if fichero.is_dir()]

    #entramos a los lotes
    for carpeta in carpetas:
        #entramos a los lotes
        with os.scandir(f'{ruta_general}/{carpeta}') as ficheros:
            carpetas_l = [fichero.name for fichero in ficheros if fichero.is_dir()]
            carpetas_l = sorted(carpetas_l)

            #checamos sus cantidades
            for carpeta_l in carpetas_l:
                with os.scandir(f'{ruta_general}/{carpeta}/{carpeta_l}') as ficheros:
                    archivos = [fichero.name for fichero in ficheros if fichero.is_file()]
                    print(f'{carpeta}: {carpeta_l}: {len(archivos)}')
            print('\n')


def mv_pics():
    carpeta_or = input("Ingresa carpeta origen: ")
    r = input(f'Esta bien este destino? {carpeta_or} y/n' )
    if r == 'y':
        #cantidad = int(input("Cantidad de archivos a mover: "))
        carpeta_des = input("Carpeta destino: ")
        with os.scandir(carpeta_or) as ficheros:
            images = [fichero.name for fichero in ficheros if fichero.is_file()]
            img_mv = images[1000:]

        for img in img_mv:
            shutil.move(f'{carpeta_or}/{img}', f'{carpeta_des}/{img}')
    else:
        mv_pics()

def rename(ruta, num):
    with os.scandir(ruta) as ficheros:
        images = [fichero.name for fichero in ficheros if fichero.is_file()]
    count = num
    for img in images:
        os.rename(f'{ruta}/{img}', f'{ruta}/jaguar_{str(count)}.jpg')
        count = count + 1
    
```
